```python
import pandas as pd 
import geopandas as gpd
import matplotlib.pyplot as plt 
```

```python
shp_gdf = gpd.read_file('2011_Dist.shp')
```

```python
shp_gdf
```

```python
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('Paramedical Staffs at District Hospitals in India as of 31st March 2019',
             fontdict={'fontsize': '15', 'fontweight' : '3'})
fig = shp_gdf.plot(column='ST_CEN_CD', cmap='RdYlGn', linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)
```
