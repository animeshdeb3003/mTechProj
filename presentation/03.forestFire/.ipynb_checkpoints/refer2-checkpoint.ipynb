## Data Loading and pre-processing

```python
shp_gdf = gpd.read_file('Census_2001/2001_Dist.shp')
```

```python
data = pd.read_csv('AHS_districtwise_2012-13.csv')
data.shape
```

```python
column_list = data.columns.to_list()
total_list = [i for i in column_list if ('urban' not in i.lower() and 'rural' not in i.lower())]
data = data.loc[:, total_list]
```

```python
x = data.iloc[:,8:]
Y = data.iloc[:,0]
dist = data.iloc[:,1]
y = np.zeros(Y.shape)
```

```python
y[np.where(Y=='Assam')] = 1
y[np.where(Y=='Bihar')] = 2
y[np.where(Y=='Chhattisgarh')] = 3
y[np.where(Y=='Jharkhand')] = 4
y[np.where(Y=='Madhya Pradesh')] = 5
y[np.where(Y=='Odisha')] = 6
y[np.where(Y=='Rajasthan')] = 7
y[np.where(Y=='Uttar Pradesh')] = 8
y[np.where(Y=='Uttarakhand')] = 9
y = y.astype('int64')
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_minmax = min_max_scaler.fit_transform(x)
```

## Applying TSNE

```python
tsne = TSNE(n_components=2, perplexity = 20, 
            n_iter=5000, method = 'exact')

tsne_result = tsne.fit_transform(X_minmax)
```

```python
tsne_result_df = pd.DataFrame({'t-SNE feature 1': tsne_result[:,0], 't-SNE feature 2': tsne_result[:,1], 
                               'label': Y})
fig, ax = plt.subplots(figsize=(12,10))
sns.scatterplot(x='t-SNE feature 1', y='t-SNE feature 2', hue='label', data=tsne_result_df, 
                ax=ax,s=120, palette = sns.color_palette("hls", 9))
lim = (tsne_result.min()-5, tsne_result.max()+5)
ax.set_xlim(lim)
ax.set_ylim(lim)
ax.set_aspect('equal')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
```

```python
kmeans = KMeans(n_clusters=14, random_state=0)
kmeans.fit(tsne_result)
```

```python
tsne_result_df = pd.DataFrame({'t-SNE feature 1': tsne_result[:,0], 't-SNE feature 2': tsne_result[:,1],
                              'label': kmeans.labels_})
fig, ax = plt.subplots(figsize=(12,10))
sns.scatterplot(x='t-SNE feature 1', y='t-SNE feature 2', hue='label', data=tsne_result_df, 
                ax=ax,s=120, palette = sns.color_palette("hls", 14))
lim = (tsne_result.min()-5, tsne_result.max()+5)
ax.set_xlim(lim)
ax.set_ylim(lim)
ax.set_aspect('equal')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
```

## Aplying Subspace Clustering

```python
R = 1 # toggle run proclus
RS = 0 # toggle use random seed

if R: # run proclus
    rseed = 1142082
    #91696 12 20 40 10
    #1142082 12 20 40 15
    if RS:
        rseed = np.random.randint(low = 0, high = 1239831)

    print "Using seed %d" % rseed

    M, D, A = prc.proclus(X_minmax, k = 12, l = 20, seed = rseed, A = 40, B = 15)
    print "Accuracy: %.4f" % prc.computeBasicAccuracy(A, dummy_y)
    print "Accuracy: %.4f"% prc.computeBasicAccuracy(A, y)
    adj = adjrand.computeAdjustedRandIndex(A, dummy_y)
    print "Adjusted rand index: %.4f" % adj
```

```python
np.unique(A, return_counts=True)
```

```python
Assign = testProclus(A, M, D, pd.DataFrame(X_minmax))
to_be_assigned = pd.DataFrame(X_minmax[np.where(A==-1)[0]])
```

```python
dummyA = np.where(A==-1)[0].tolist()
for i in range(to_be_assigned.shape[0]):
    A[dummyA[i]] = Assign.assignCluster(to_be_assigned.iloc[i,:])
```
