```python
X = X.iloc[:,1:]
X['sex'] = pd.factorize(X['sex'])[0].reshape((395,1))
X['address'] = pd.factorize(X['address'])[0].reshape((395,1))
X['famsize'] = pd.factorize(X['famsize'])[0].reshape((395,1))
X['Pstatus'] = pd.factorize(X['Pstatus'])[0].reshape((395,1))
X['schoolsup'] = pd.factorize(X['schoolsup'])[0].reshape((395,1))
X['famsup'] = pd.factorize(X['famsup'])[0].reshape((395,1))
X['paid'] = pd.factorize(X['paid'])[0].reshape((395,1))
X['activities'] = pd.factorize(X['activities'])[0].reshape((395,1))
X['nursery'] = pd.factorize(X['nursery'])[0].reshape((395,1))
X['higher'] = pd.factorize(X['higher'])[0].reshape((395,1))
X['internet'] = pd.factorize(X['internet'])[0].reshape((395,1))
X['romantic'] = pd.factorize(X['romantic'])[0].reshape((395,1))
```

```python
X = pd.concat([X,pd.get_dummies(X['Mjob'], prefix = 'Mjob')], axis = 1)
X = pd.concat([X,pd.get_dummies(X['Fjob'], prefix = 'Fjob')], axis = 1)
X = pd.concat([X,pd.get_dummies(X['reason'], prefix = 'reason')], axis = 1)
X = pd.concat([X,pd.get_dummies(X['guardian'], prefix = 'guardian')], axis = 1)
```

```python
X.drop(['Mjob', 'Fjob', 'reason', 'guardian'], axis = 1, inplace = True)
```

```python
tsne = TSNE(n_components=2, perplexity = 50) 
            #n_iter=5000, method = 'exact')

tsne_result = tsne.fit_transform(X_minmax)
```

```python
tsne_result_df = pd.DataFrame({'t-SNE feature 1': tsne_result[:,0], 't-SNE feature 2': tsne_result[:,1],'Class': Y})
fig, ax = plt.subplots(figsize=(12,10))
sns.scatterplot(x='t-SNE feature 1', y='t-SNE feature 2', hue='Class', data=tsne_result_df, 
                ax=ax,s=120, palette = sns.color_palette("hls", 4))
lim = (tsne_result.min()-5, tsne_result.max()+5)
ax.set_xlim(lim)
ax.set_ylim(lim)
ax.set_aspect('equal')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(tsne_result)
```

```python
tsne_result_df = pd.DataFrame({'t-SNE feature 1': tsne_result[:,0], 't-SNE feature 2': tsne_result[:,1],
                              'label': kmeans.labels_})
fig, ax = plt.subplots(figsize=(12,10))
sns.scatterplot(x='t-SNE feature 1', y='t-SNE feature 2', hue='label', data=tsne_result_df, 
                ax=ax,s=120, palette = sns.color_palette("hls", 3))
lim = (tsne_result.min()-5, tsne_result.max()+5)
ax.set_xlim(lim)
ax.set_ylim(lim)
ax.set_aspect('equal')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
```
